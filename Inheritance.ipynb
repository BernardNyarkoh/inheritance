Simple Inheritance


```python
class Person:
    def greet(self):
        print("Good day everyone!")

class Student(Person):
    pass

s = Student()
s.greet()
```

Inheriting Attributes

```python
class Person:
    def __init__(self,name):
        self.name = name

class Student(Person):
    pass

s = Student("Matthew")
print(s.name)
```

Extending a parent class

```python
class Person:
    def greet(self):
        print("Hello")

class Student(Person):
    def study(self):
        print("I'm enjoying the python OOP Course")

daniella  = Student()

daniella.greet()
daniella.study()
```

Constructor Inheritance

```python
class Person:
    def __init__(self,name):
        self.name = name

class Student(Person):
    def __init__(self,name,student_id):
        super().__init__(name)
        self.student_id = student_id

student = Student("Matthew","S12345")
print(student.student_id)
print(student.name)
```

Method Overriding

```python
class Anmial:
    def speak(self):
        print("Animal makes a sound")

class Dog(Anmial):
    def speak(self):
        print("Dog barks")

dog = Dog()
dog.speak()
```

Calling The Parent Method

```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")
```
